## INSTALL RASQUAL

### Install CLAPACK

```python
cd ~/run_rasqual/
```

```python
! mkdir CLAPACK; cd CLAPACK; wget http://www.netlib.org/clapack/clapack-3.2.1-CMAKE.tgz
```

```python
cd CLAPACK/
```

```python
!tar zxvf clapack-3.2.1-CMAKE.tgz
```

```python
cd clapack-3.2.1-CMAKE/
```

```python
! mv make.inc.example make.inc; ulimit -s unlimited; make # note: if no ulimit command, it will fail due to segfault.
```

```python
!ln -s lapack_LINUX.a liblapack.a
!ln -s tmglib_LINUX.a libtmglib.a
!ln -s blas_LINUX.a libblas.a
```
